```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

# Load CSV files directly from Drive
trader = pd.read_csv('/content/drive/MyDrive/ds_Kavya_Bhardwaj/csv_files/historical_data.csv')
sentiment = pd.read_csv('/content/drive/MyDrive/ds_Kavya_Bhardwaj/csv_files/fear_greed_index.csv')
```

```python
# Convert timestamps to date
trader['date'] = pd.to_datetime(trader['Timestamp IST'], format="%d-%m-%Y %H:%M").dt.date
sentiment['date'] = pd.to_datetime(sentiment['new_timestamp'], format="%d-%m-%Y %H:%M").dt.date

# Merge both datasets
merged = pd.merge(trader, sentiment, on='date', how='left')

# Display results
print(merged.shape)
print(merged[['date', 'value', 'classification']].head())
```

```python
output_path = '/content/drive/MyDrive/ds_Kavya_Bhardwaj/csv_files/merged_trader_sentiment.csv'
merged.to_csv(output_path, index=False)
print(f"Merged file at: {output_path}")
```
